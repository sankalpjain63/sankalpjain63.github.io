### Visualization 1:  Using a line chart, visualize the average square footage of buildings constructed from 1800 to 2022

```python
#generating plot 1

Viz1 = alt.Chart.from_dict({
  "data": {
    "values": [{"x":yearcons, "y":sqfoot}],
  },
  "selection":{"brush":{"type":"interval"}},
  "transform":[
    {"flatten":["x","y"]},
    {"calculate":"toDate(datum.x)", "as":"x"}
  ], 
  "mark":{"type":"line", "tooltip":True},
  "encoding":{
    "x":{"field":"x", "type":"temporal", 
        "axis": {
          "title" : "Year"
        }},
    "y":{"aggregate": "mean", "field":"y", "type":"quantitative",
      "axis": {
        "title": "Average Square Footage"
      }},
    "color":{"condition":{"selection":"brush", "value": "blue"}}
  }
}
).properties(
    width=600,
    height=400,
    title="Average Square Footage VS Year")
```

### Write-up visualization 1

### The goal of this visualization is to examine the changes in the average size of buildings over time, as the reduction of open spaces has resulted in smaller buildings. To achieve this, we utilized temporal encoding for the X-axis to map the year values and calculated the average square footage for the Y-axis from the dataframe. We removed rows with empty values in the year constructed and square footage fields and filtered the data to only include buildings constructed between 1800 and 2022. We utilized the ".properties" trait in Altair to encode the width and height of the chart. The original color of the line chart is blue.

## Visualization 2 with Interactivity: Interactive bar chart shows total building square footage by status for each city in dataset. Users can filter for specific city. 

### Citation: "https://stackoverflow.com/questions/60964087/how-can-i-select-data-for-a-chart-based-on-a-dropdown"

```python
#generating plot 2
#data = inventory2.copy()

select_box = alt.binding_select(options=list(df4['City'].unique()))
selection = alt.selection_single(fields=['City'], bind=select_box)
    
Viz2 = alt.Chart(df4).mark_bar(tooltip = True).encode(
    x = alt.X('Bldg Status:O', axis = alt.Axis(title = 'Building Status') ,),
    y = alt.Y('sum(Square Footage):Q', axis = alt.Axis(title = 'Total Square Footage',), ),
).properties(
    title = "Total Square Footage VS Building Status",
    width = 600,
    height = 400,
).add_selection(
    selection
).transform_filter(
    selection
)
```
